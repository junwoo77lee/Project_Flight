```python
airport_data = list(pd.read_html('https://en.wikipedia.org/wiki/List_of_the_busiest_airports_in_the_United_States'))
```

```python
airport_rank_1_to_29 = airport_data[0]
airport_rank_1_to_29.to_csv("first.csv")
```

```python
airport_rank_31_to_61 = airport_data[1]
airport_rank_31_to_61.to_csv("second.csv")
```

```python
full_data_set = [airport_rank_1_to_29, airport_rank_31_to_61]
result = pd.concat(full_data_set)
result
```
